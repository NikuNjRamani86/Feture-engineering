# 4. Traget guided ordinal encoding

```python
df = pd.read_csv('titanic_train.csv')
```

```python
df.head()
```

```python
df['cabin'].fillna('missing')
```

```python
data = df[['cabin', 'survived']]
```

```python
data['cabin'].fillna('missing', inplace=True)
```

```python
data.head()
```

```python
data['cabin'] = data['cabin'].astype(str).str[0]
```

```python
data.cabin.unique()
```

```python
data.groupby(['cabin'])['survived'].mean()
```

```python
data.groupby(['cabin'])['survived'].mean().sort_values().index
```

```python
label = data.groupby(['cabin'])['survived'].mean().sort_values().index
```

```python
enumerate(label, 0)
```

```python
label2 = {k:i for i,k in enumerate(label,0)}
```

```python
label2
```

```python
data['label']=data['cabin'].map(label2)
```
